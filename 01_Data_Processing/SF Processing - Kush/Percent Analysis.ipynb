```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("~/Documents/CCI/DS_Projects/Twitter/Percent_for_Overall_SF.csv")
#To avoid skewed percentages only take tracts that have high samples
data = data[data["total"] > 50]
data = data.drop("Unnamed: 0" , axis=1)
```

```python
data.head()
```

```python
def plotColratio(col1 , col2):
    plt.hist(data[col2] / (data[col1]+ 0.000001) , bins = np.arange(0,5,0.1))
    plt.xlim([0,5])
```

```python
night = data["home night"] + data["neighbor night"] + data["non-neighbor night"]
```

```python
day = data["home day"] + data["neighbor day"] + data["non-neighbor day"]
```

```python
print("Day Tweet total percentage: " , sum(day))
print("Day Tweet total percentage: " , sum(night))
```

```python
plt.hist(day/night)
```

## Looking at various columns against eachother and their ratios in the format (y/x) to see what is more common 

```python
data.plot.scatter("home" , "non-neighbor")
plt.show()
plotColratio("home" ,"non-neighbor")
```
